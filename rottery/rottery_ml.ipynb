```python
display(rottery_uncertain)
display(rottery_certain)
```

```python
rottery_dataset = pandas.concat([rottery_certain, rottery_uncertain], axis=0)
rottery_dataset
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import *
```

```python
target = 'result'
x = rottery_dataset.drop(target, axis=1)
y = rottery_dataset.loc[:, target]
display(x)
display(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1)
```

```python


model_xgb = XGBClassifier(max_depth=3)
model_xgb.fit(x_train, y_train)
y_pred = model_xgb.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
model_lgbm = LGBMClassifier(max_depth=3)
model_lgbm.fit(x_train, y_train)
y_pred = model_lgbm.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
model = CatBoostClassifier(
    iterations=1000,        # 트리 수 (early stopping 때문에 많게)
    learning_rate=0.1,      # 학습률
    depth=6,                # 트리 깊이
    eval_metric='Accuracy', # 평가 지표
    verbose=100,            # 100 iteration마다 로그 출력
    random_seed=42
)

# 학습
model.fit(
    x_train, y_train,
    eval_set=(x_test, y_test),
    early_stopping_rounds=50  # 50번 동안 개선 없으면 조기종료
)

# 5. 예측
y_pred = model.predict(x_test)

# 6. 평가
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
